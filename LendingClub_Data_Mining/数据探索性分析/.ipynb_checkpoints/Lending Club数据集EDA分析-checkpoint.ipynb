# 项目介绍

所谓探索性数据分析(Exploratory Data Analysis，以下简称EDA)，是指对已有的数据(特别是调查或观察得来的原始数据)在尽量少的先验假定下进行探索，通过作图、制表、方程拟合、计算特征量等手段探索数据的结构和规律的一种数据分析方法。特别是当我们对这些数据中的信息没有足够的经验，不知道该用何种传统统计方法进行分析时，探索性数据分析就会非常有效。探索性数据分析在上世纪六十年代被提出，其方法由美国著名统计学家约翰·图基(John Tukey)命名。

# 项目背景

Lending Club 是全球最大的撮合借款人和投资人的线上金融平台，它利用互联网模式建立了一种比传统银行系统更有效率的、能够在借款人和投资人之间自由配置资本的机制。
通过Lending Club 的平台，借款人和小微企业可以获得更低的利率，投资人则可以获得较好的收益。

公司于2006 年10 月在特拉华州以“SocBank”的名字成立，之后于2006 年11 月更名为Lending Club。

截至 2014 年6 月30 日，Lending Club 累计贷款金额超过50 亿美元，支付给投资人的利息超过4.9 亿美元。

### 处理缺失值

统计每列属性缺失值的数量

```python
def not_null_count(column):
    column_null = pd.isnull(column)
    null = column[column_null]
    return len(null)

column_null_count = data.apply(not_null_count)
print(column_null_count)
```

```python
half_count = len(data)/2 #设定阈值
data = data.dropna(thresh=half_count,axis=1)# 若某一列数据缺失的数量超过阈值就被删除
data.to_csv('loan_2018q1.csv',index=False)
```

### 单变量分析
+ 1.贷款状态分布

### 处理异常值

```python
used_data[used_data['loan_status'].isnull()]
```

由于loan_status异常值数量较小，因此我们直接删异常值所对应的行

```python
used_data = used_data.dropna()
used_data['loan_status'].value_counts()
```

为了方便分析，我们将贷款状态进行分类变量编码，主要将贷款状态分为正常和违约

```python
codeDict={'Current':0,'Fully Paid':0,'Late (31-120 days)':1,'In Grace Period':1,'Late (16-30 days)':1,'Charged Off':1,'Default':1}
used_data['loan_status_coded'] = used_data['loan_status'].map(codeDict)
print('After Coding:')
used_data['loan_status_coded'].value_counts()
```

```python
loan_status_values = pd.value_counts(used_data['loan_status_coded']).values
labels = ['正常','违约']
plt.figure(figsize=(6,6))
plt.pie(loan_status_values,explode = [0.1,0.1],
        labels = labels, shadow = True,autopct='%1.2f%%',
        colors = ['c', 'r'])
plt.title('贷款状态占比')
plt.legend()
plt.show()
```

从图中可以看出，平台贷款发生违约的数量占少数。贷款状态为正常的有105804个，贷款正常状态占比为98.09%。贷款状态将作为我们建模的标签，贷款状态正常和贷款状态违约两者数量不平衡，绝大多数常见的机器学习算法对于不平衡数据集都不能很好地工作,所以后续建模时要解决样本不均衡问题。

```python
mapping_dict = {
    "emp_length": {
        "10+ years": 10,
        "9 years": 9,
        "8 years": 8,
        "7 years": 7,
        "6 years": 6,
        "5 years": 5,
        "4 years": 4,
        "3 years": 3,
        "2 years": 2,
        "1 year": 1,
        "< 1 year": 0,
    }
}
used_data = used_data.replace(mapping_dict)
used_data['annual_inc'] = used_data['annual_inc'].astype('float').dropna()
```

## 总结

```python
used_data.corr()#计算相关系数2
```

```python
# 相关系数图
used_data = used_data.drop(['delinq_2yrs'],axis=1)
names = ['loan_amnt', 'annual_inc' ,'emp_length', 'Loan_Status_Coded','int_rate' ] #设置变量名
correlations = used_data.corr()
# plot correlation matrix
plt.figure(figsize=(19, 9))
fig = plt.figure() #调用figure创建一个绘图对象
ax = fig.add_subplot(111)
cax = sns.heatmap(correlations, vmin=-1, vmax=1)  #绘制热力图，从-1到1
ticks = np.arange(0,5,1) #生成0-5，步长为1
ax.set_xticks(ticks)  #生成刻度
ax.set_yticks(ticks)
ax.set_xticklabels(names) #生成x轴标签
ax.set_yticklabels(names)
plt.xticks(rotation=90)
fig.savefig("Corr")
plt.show()
```

### 1.影响风险的因素

高收入客户意味着有良好的现金流，偿还债务能力较高，违约可能性较低，一般来说，这样的客户信用评级也相对较高，平台对应的贷款风险也较低；长期有不良信用记录的客户再次发生违约的可能性较高，对此类用户，应加大风险防范措施。

### 2.Lending Club平台特点

+ 平台业务持续稳定发展：第一季度总体增长，平台主要业务主要集中在加州，德克萨斯州和纽约州。
+ 平台贷款金额以小额贷款为主，主要集中在10000美元左右。
+ 平台利率较高，集中在12.62%左右，贷款利率相对传统金融机构较高。
+ 平台一季度违约风险低，平台发生违约数量较少，贷款正常占比为98.19%。

### 3.个人建议

+ 完善客户画像和产品设计：信贷业务开展前，首先明确信贷机构的目标客户群，目标客户的特征和画像信息是什么。
+ 优化贷款模型：完善客户信息的同时，借助机器学习的技术持续优化贷款模型。
